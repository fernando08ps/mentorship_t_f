```python
import pandas as pd

# Load the dataset
movies = pd.read_csv("https://raw.githubusercontent.com/CTopham/TophamRepo/master/Movie%20Project/Resources/tmdb_5000_movies.csv")

# Remove the specified columns
columns_to_remove = ['homepage', 'keywords', 'overview', 'tagline']
movies = movies.drop(columns=columns_to_remove, axis=1)

# Display the updated DataFrame
print(movies.head())
```

### CORRELATIONS

```python
movies.corr()
```

```python
# Correlation between all the columns of DataFrame.
correlation_matrix = movies.corr()

# Create a heatmap of the correlation matrix
plt.matshow(correlation_matrix, cmap='Blues', vmin=-1, vmax=1)

# Add labels to the heatmap
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        value = correlation_matrix.iloc[i, j]
        text = "{:.2f}".format(value)
        if value > 0:
            color = 'black'
        else:
            color = 'white'
        plt.text(j, i, text, ha="center", va="center", color=color, fontsize=8)

plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.show()
```

**Based on the correlation coefficients presented, the following conclusions can be drawn:**

Budget and revenue are positively correlated, with a correlation coefficient of 0.730823. This indicates that movies with higher budgets tend to have higher revenue.

Budget and runtime are also positively correlated, with a correlation coefficient of 0.269851. This indicates that movies with higher budgets tend to have longer runtimes.

Popularity and revenue are positively correlated, with a correlation coefficient of 0.644724. This indicates that movies that are more popular tend to have higher revenue.

Popularity and runtime are also positively correlated, with a correlation coefficient of 0.225502. This indicates that movies that are more popular tend to have longer runtimes.

Vote average and vote count are positively correlated, with a correlation coefficient of 0.778130. This indicates that movies that receive more votes tend to have higher average ratings.

Note that all of these correlations are significant at the 0.05 level.

**Interpreting the correlations**

The positive correlations between budget and revenue, popularity and revenue, and popularity and runtime suggest that these factors are all important determinants of movie success. Movies with higher budgets, more popular appeal, and longer runtimes tend to be more successful in terms of revenue.

The positive correlation between vote average and vote count suggests that movies that are more popular tend to receive higher average ratings from viewers. This is likely because more people are exposed to popular movies, and therefore have more opportunities to rate them.

**Limitations**

It is important to note that these correlations are based on a limited dataset of movies. It is possible that the correlations would be different if the dataset were larger or more representative.

Additionally, the correlations only show that there is a relationship between the variables. They do not prove that one variable causes the other. For example, it is possible that movies with higher budgets are simply more likely to be popular, rather than the other way around.

BUDGET AND REVENUE PER GENRE (MEDIAN)

```python
import pandas as pd

# Load the dataset
movies = pd.read_csv("https://raw.githubusercontent.com/CTopham/TophamRepo/master/Movie%20Project/Resources/tmdb_5000_movies.csv")

# Explode the 'genres' column to have one genre per row
movies['genres'] = movies['genres'].apply(lambda x: [genre['name'] for genre in eval(x)])
movies = movies.explode('genres')

# Group by genre and calculate the median budget and revenue
genre_median_data = movies.groupby('genres').agg({'budget': 'median', 'revenue': 'median'}).reset_index()

# Display the genres with the highest median budget and revenue
print("Genres with the highest median budget and revenue:")
print(genre_median_data.sort_values(by='budget', ascending=False))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Load the dataset
movies = pd.read_csv("https://raw.githubusercontent.com/CTopham/TophamRepo/master/Movie%20Project/Resources/tmdb_5000_movies.csv")

# Explode the 'genres' column to have one genre per row
movies['genres'] = movies['genres'].apply(lambda x: [genre['name'] for genre in eval(x)])
movies = movies.explode('genres')

# Group by genre and calculate the median budget and revenue
genre_median_data = movies.groupby('genres').agg({'budget': 'median', 'revenue': 'median'}).reset_index()

# Sort the data by median budget for a better visualization
sorted_data = genre_median_data.sort_values(by='budget', ascending=False)

# Set the plot style
sns.set(style="whitegrid")

# Create a bar plot for median budget
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='budget', y='genres', data=sorted_data, palette='viridis')

# Format x-axis labels with one decimal place for tens of millions
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f"${x/1e6:.0f}M"))

plt.title('Genres with the Highest Median Budget')
plt.xlabel('Median Budget')
plt.ylabel('Genres')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

### POPULAR GENRES AND MEDIAN REVENUES

```python
# Calculate median popularity and revenue by genre
genre_popularity = movies.groupby('genres')['popularity'].median()
genre_revenue = movies.groupby('genres')['revenue'].median()

# Identify top genres by popularity and revenue
top_popular_genres = genre_popularity.nlargest(5)
top_revenue_genres = genre_revenue.nlargest(5)

# Create bar plots for top genres by popularity and revenue
plt.figure(figsize=(12, 6))

# Top Genres by Median Popularity
plt.subplot(1, 2, 1)
sns.barplot(x=top_popular_genres.values, y=top_popular_genres.index, palette='viridis')
plt.title('Top 5 Genres by Median Popularity')
plt.xlabel('Median Popularity')
plt.ylabel('Genres')

# Top Genres by Median Revenue
plt.subplot(1, 2, 2)
sns.barplot(x=top_revenue_genres.values, y=top_revenue_genres.index, palette='viridis')
plt.title('Top 5 Genres by Median Revenue')
plt.xlabel('Median Revenue')
plt.ylabel('Genres')

plt.tight_layout()
plt.show()
```
